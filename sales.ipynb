```python
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df_my = df.groupby('month_year')['sales'].sum().reset_index()
df_my
```

```python
plt.figure(figsize=(16, 5))
plt.plot(df_my['month_year'], df_my['sales'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

```python
p_s = df.groupby(['month','year'])['sales'].agg(['sum','count']).sort_values(by='month').reset_index()
p_s['month'] = pd.to_datetime(p_s['month'], format='%m').dt.month_name()
p_s
```

```python
plt.figure(figsize=(16, 10))
sns.scatterplot(data=p_s, x="count", y="sum", size='year', hue="month")
plt.show()
```

```python
plt.figure(figsize=(16, 10))
sns.barplot(data=p_s, x="month", y="count", hue="year")
plt.show()
```
